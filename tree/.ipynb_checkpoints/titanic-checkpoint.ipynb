```python

import pandas as pd
```

```python
df = pd.read_csv('titanic_dataset.csv',delimiter=';')
```

```python
df.head()
```

```python
# lets clear our data for so that we do not have to use not required value
inputs = df.drop(['Survived','PassengerId','Name','SibSp','Parch','Cabin','Embarked','Ticket'],axis='columns')
```

```python
inputs
```

```python
# we need to encode the value of sex to the numeric so lets encode it
from sklearn.preprocessing import LabelEncoder
```

```python
# create the instance of labelEncoder
encoder = LabelEncoder()
```

```python
# lets apply the encoder to our column
inputs['Sex_n'] = encoder.fit_transform(inputs.Sex)
```

```python
# lets remove the Sex column with string data
dfnew = inputs.drop('Sex',axis='columns')
dfnew
```

```python
# in our dataset there is one column with NaN value in string format
# we need to replace it or define as 0 or our as per value
# so that we we need numpy for it
import numpy as np
```

```python
#replace Nan Value of data with 0
datanew = dfnew.replace(np.NaN,0)
```

```python
datanew
```

```python
# now our target is to check the survived person with category so lets define it
target = df.Survived
```

```python
# Now to make decision in our model we need to import tree from sklearn
from sklearn import tree
```

```python
# from tree lets instanciate the decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#fit our model because the model requires the dependant list of values to the target
model.fit(datanew,target)
```

```python
# now lets check the score of the model
model.score(datanew,target)
```

```python
# lets predict one ramdom person
model.predict([[1,25,40,1]])
```
